```python
import requests
import time

planets = []

for i in range(50):
    url = f"https://swapi.dev/api/planets/{i+1}"
    response = requests.get(url)

    if response.status_code == 200:
        json_data = response.json()
        planets.append(
            (json_data["name"],
             json_data["population"],
             json_data["climate"],
             json_data["terrain"],
             json_data["gravity"],
             json_data["residents"])
        )
    else:
        planets.append("error")

    time.sleep(1)

print(planets)
```

```python
import pandas as pd

planets = pd.DataFrame(planets)
list(planets.columns)
planets.columns = ["Name","Popualtion","Climate","Terrain","Gravity","Residents"]

planets
```

```python
planets.to_csv("planets.csv")
```
