```python
import requests

import html
## Web scraping basics

url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
headers = {"Accept-Language": "en-US"}
html = requests.get(url,headers = headers)
imdb = Soup(html.text)
```

```python
# list comprehension
titles = imdb.find("h3",{"class": "lister-item-header"})
clean_titles = [title.strip() for title in titles]

clean_titles
```

```python
## get rating from a website
## div: ratings-imdb-rating

ratings = imdb.find("div",{"class":"ratings-imdb-rating"})
clean_ratings = [float(rating.strip()) for rating in ratings]

clean_ratings
```

```python
certificates = imdb.find("span",{"class":"certificate"})
clean_certificates = [(cer.strip()) for cer in certificates]

clean_certificates
```

```python
length = imdb.find("span",{"class":"runtime"})
clean_length = [(leng.strip()) for leng in length]

clean_length
```

```python
genre = imdb.find("span",{"class":"genre"})
clean_genre = [(gen.strip()) for gen in genre]

clean_genre
```

```python
import pandas as pd

movies_database = pd.DataFrame(data = {
    "Titles" : clean_titles,
    "Rating" : clean_certificates,
    "Length_Minute" : clean_length,
    "Genre"  : clean_genre,
    "Score"  : clean_ratings
})

movies_database = movies_database.reset_index()
movies_database.head()
```

```python
# Create Launced Year column in a dataframe

""" extract year from name's titles """
movies_database["Year"] = movies_database["Titles"].str.extract(r'\((\d{4})\)')
""" remove year from name's titles """
movies_database["Titles"] = movies_database["Titles"].str.replace(r'\(\d{4}\)', '').str.strip()
""" remove number from name's titles """
movies_database["Titles"] = movies_database["Titles"].str.replace(r'^\d+\.\s', '')
""" remove "min" and change str to int """
movies_database["Length_Minute"] = movies_database["Length_Minute"].str.replace(' min', '').astype(int)

movies_database.head()
```

```python
colnames = list(movies_database.columns)
colnames[0] = "Number"
movies_database.columns = colnames

movies_database
```

```python
movies_database.to_csv("movies_database.csv")
```
